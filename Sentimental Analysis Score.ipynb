```python

import pandas as pd
import re
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')
nltk.download('stopwords')

from nltk import word_tokenize
from collections import Counter
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import numpy as np
from nltk.corpus import stopwords


```

```python
#Checking for a single review
pd.set_option('display.max_columns', 500)

df = pd.read_excel("E:/Ramada Plaza by Wyndham Waikiki.xlsx")

```

```python
tex = df["Customer_Review"]

tex=tex.dropna(axis=0)
tex.head(10)
```

```python
Adj_List = []
Noun_List = []
sennum = -1
analyser = SentimentIntensityAnalyzer()
w = pd.DataFrame(columns=['Noun1','Noun2','Noun3','Adj1','Adj2','Adj3','Adj4','Adj5'])
for review in tex:
    review = review.lower()
    token = word_tokenize(review)
    PoS = nltk.pos_tag(token)
    w_new = pd.DataFrame(columns=['Noun1','Noun2','Noun3','Adj1','Adj2','Adj3','Adj4','Adj5'])
    w_reset = pd.DataFrame(columns=['Noun1','Noun2','Noun3','Adj1','Adj2','Adj3','Adj4','Adj5'])
    adjN = 0
    nN = 0
    sent = nltk.sent_tokenize(review)
    for sen in sent:
        sennum = sennum + 1
        words = word_tokenize(sen)
        PoS = nltk.pos_tag(words)
        
        for word in PoS:
            if word[1] == ".":
                w = pd.concat([w,w_new],axis = 0)
                w_new = w_reset.copy()
                adjN = 0
                nN = 0
            if word[1] == "JJ":
                adjN = adjN + 1
                columnname = "Adj{}".format(adjN)
                w_new.loc[sennum,columnname] = word[0]
            
            if (word[1] == "NN" or word[1] == "NNS"):
                nN = nN + 1
                columnname = "Noun{}".format(nN)
                w_new.loc[sennum,columnname] = word[0]
                w_new.loc[sennum,'Compound'] = analyser.polarity_scores(sen).get('compound')
            
w.head(10)
```

```python
list1=list(w.columns)
Noun_list=[i for i in list1 if "Noun" in i] 
length_Noun=len(Noun_list)
w_list=[]

for i in range(length_Noun):
    other_cols=[j for j in list1 if j not in Noun_list]
    w_list.append(w.loc[:,other_cols])
    w_list[i]["Noun"]=w["Noun"+str(i+1)]
    
    
    
```

```python


w1=w_list[0]
for a in w_list[1:]:
       w1 = pd.concat([w1,a],axis = 0) 
w1
```

```python
wfin = w1.dropna(axis=0, subset=['Noun'])

wfin
```

```python
list1=list(wfin.columns)
Adj_list=[i for i in list1 if "Adj" in i] 
length_Adj=len(Adj_list)
w_list=[]

for i in range(length_Adj):
    other_cols=[j for j in list1 if j not in Adj_list]
    w_list.append(wfin.loc[:,other_cols])
    w_list[i]["Adj"]=wfin["Adj"+str(i+1)]
    
    
    
```

```python

w1=w_list[0]
for a in w_list[1:]:
       w1 = pd.concat([w1,a],axis = 0) 

w1
```

```python
wfin1 = w1.dropna(axis=0, subset=['Adj'])

#wfin1
bool_list=[]
for i in wfin1['Noun'].values:
    bool_list.append(i not in stopwords.words('english'))
wfin1=wfin1[bool_list]
bool_list=[]
for i in wfin1['Adj'].values:
    bool_list.append(i not in stopwords.words('english'))
wfin1=wfin1[bool_list]
wfin1.head(10)
```

```python
AdjNoun = wfin1.reset_index()

AdjNoun.head(10)
```

```python
freqNouns = wfin1.groupby(['Noun']).count().sort_values(['Adj'], ascending=False).head(60)

freqNouns.head(10)
```

```python
index = []
for i in freqNouns.iterrows():
    index.append(i[0])
print(index)
```
